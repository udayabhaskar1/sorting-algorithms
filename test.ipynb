```python
def count(data, target):
    n = 0
    for item in data:
        if (item == target):
            n += 1
    return n
```

### Finding the factors of n

```python
def factors(n):
    results = []
    for k in range(1, n+1):
        if (n % k == 0):
            results.append(k)
    return results
```

```python
factors(9)
```

#### Generator

```python
def factors(n):
    results = []
    for k in range(1, n+1):
        if (n%k == 0):
            yield k
```

```python
factors(9)
```

```python
from time import time
start_time = time()
factors(100000000000000000)
end_time = time()
elapsed = end_time - start_time
```

```python
def if_else(n):
    n = int(input().strip())
    if (n % 2 != 0):
        print("Weird")
    else:
        if (n >=2 & n <= 5):
            print("Not Weird")
        elif (6 <= n <= 20):
            print("Weird")
        else:
            print("Not Weird")
```
